```python
import numpy as np
import matplotlib.pyplot as plt
import json
```

```python
with open('stats.json') as f:
    stats = json.load(f)
    
type_masks = {}
for i, n in enumerate(stats):
    type = stats[n]['type']
    if type not in type_masks:
        type_masks[type] = [False] * len(stats)
    type_masks[type][i] = True
```

```python
for type, mask in type_masks.items():
    indices = np.arange(len(stats))[mask]
    memory = np.array([v['memory'] for v in stats.values()])[mask]
    plt.plot(indices, memory, label=type)
plt.legend()
```

```python
for type, mask in type_masks.items():
    indices = np.arange(len(stats))[mask]
    memory = np.array([v['memory'] for v in stats.values()])[mask]
    memory_diff = np.ndarray(memory.shape)
    memory_diff[0] = 0
    memory_diff[1:] = memory[1:] - memory[:-1]
    plt.plot(indices, memory_diff, label=type)
plt.ylim(bottom=-100000, top=100000)
plt.legend()
```
