```python
import numpy as np
import matplotlib.pyplot as plt
import json

with open('stats.json') as f:
    stats = json.load(f)
```

```python
plt.plot([0,1,2,3,4], [1,2,None,9,11])
```

```python
type_masks = {}
for i, n in enumerate(stats):
    type = stats[n]['type']
    if type not in type_masks:
        type_masks[type] = [False] * len(stats)
    type_masks[type][i] = True
```

```python
np.arange(10)
```

```python
for type, mask in type_masks.items():
    indices = np.arange(len(stats))[mask]
    memory = np.array([v['memory'] for v in stats.values()])[mask]
    plt.plot(indices, memory, label=type)
plt.legend()
```
